```python
x = np.array([0.9, 0.8 , 0.7 , 0.6 , 0.5 , 0.4 , 0.3 , 0.2 , 0.1])
```

```python
y = np.array([0.13, 0.15, 0.23, 0.25,0.33, 0.35, 0.43, 0.45, 0.49])
```

```python
plt.scatter(x,y, alpha= 0.3)
plt.show
```

```python
#we add a column of 1s for the indendent term
x_ = np.array([np.ones(x.size), x]).T
```

#Minimum mean square error

$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
# matrix multiplication
B = np.linalg.inv(x_.T @ x_) @ x_.T @ y
```

```python
B
```

```python
plt.scatter(x,y, alpha= 0.3)
plt.plot([0.1e7, 3.5e7], [B[0] + B[1] * 0.1e7, B[0] + B[1] * 3.5e7], c= "red")
plt.show
```
